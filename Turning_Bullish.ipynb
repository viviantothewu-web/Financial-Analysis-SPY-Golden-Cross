```python
# finding golden cross

import pandas as pd

spy_df = pd.read_csv("SPY_close_price_5Y.csv", parse_dates = ["Date"])

spy_df.head()
```

```python
spy_df.assign( # add new columns in existing dataframe.

    # adding moving average 50 column
    # d.close selects the closing price column
    # window --> average over 50 data points
    # d is just a placeholder name for the spy_df, could have written lambda df:
    moving_average_50 = lambda d: d.Close.rolling(window = 50, min_periods=50).mean(),

    # adding moving average 200 column
    moving_average_200 = lambda d: d.Close.rolling(window = 200, min_periods = 200).mean(),

    # lambda for golden cross
    golden_cross = lambda d: ((d.moving_average_50 > d.moving_average_200) & #short term MA is higher than long term MA 
                              # On the previous day (.shift(1)), the short-term average (MA50) was <= to MA200.
                              (d.moving_average_50.shift(1) <= d.moving_average_200.shift(1))).astype(int)
).query("golden_cross ==1")
```
